```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
laliga_df = pd.read_csv("C:\\Users\\si00467932\\\Downloads\\Laliga.csv",skiprows=1)
```

```python
#Replace dash with zero
#laliga_df[laliga_df.iloc[0:61,0:20]== '-'] = '0'
laliga_df.replace('-',0,inplace=True)
```

```python
#Print all the teams which have started playing between 1930-1980. 
a=list(map(lambda x:(x[0:4]>='1930' and x[0:4]<='1980') and (x[5:7]>='30' and x[5:7]<='80')
              ,laliga_df['Debut']))
for i,x in enumerate(a):
    if x==True:
        print(laliga_df.iloc[i]['Team'],' - ',laliga_df.iloc[i]['Debut'])
```

```python
#list of teams which came Top 5 in terms of points
laliga_df['Points']=list(map(lambda x:int(x),laliga_df['Points'].values))
laliga_df.sort_values(by='Points',ascending=False,inplace=True)
print(laliga_df.iloc[0:5]['Team'])
```

```python
laliga_df['GoalsFor']=list(map(lambda x:int(x),laliga_df['GoalsFor'].values))
laliga_df['GoalsAgainst']=list(map(lambda x:int(x),laliga_df['GoalsAgainst'].values))

def Goal_diff_count(df):
    goal_diff_df = pd.DataFrame(columns=['Team', 'Goal Difference'])
    goal_diff_df['Team'] = df['Team']
    goal_diff_df['Goal Difference'] = abs(df['GoalsFor']-df['GoalsAgainst'])
    return goal_diff_df,goal_diff_df.iloc[goal_diff_df['Goal Difference'].idxmax()]['Team'],goal_diff_df.iloc[goal_diff_df['Goal Difference'].idxmin()]['Team']

    
goal_diff, max_tem, min_team = Goal_diff_count(laliga_df)
print(goal_diff)
print("Team with max goal Difference is",max_team)
print("Team with min goal Difference is",min_team)
```

```python
# new column with the name “Winning Percent” and append it to the data set 

#laliga_df['GamesWon'] = laliga_df['GamesWon'].astype('int64')
laliga_df['GamesWon'] = list(map(lambda x:int(x),laliga_df['GamesWon'].values))
laliga_df['GamesPlayed'] = list(map(lambda x:int(x),laliga_df['GamesPlayed'].values))
laliga_df['Winning Percent']=(laliga_df['GamesWon']/laliga_df['GamesPlayed']) * 100
```

```python
laliga_df.columns.values
```

```python
#Print the top 5 teams which have the highest Winning percentage

#print(laliga_df.iloc[0:61]['Winning Percent'])
#replace NaN values of winning percent to 0
laliga_df['Winning Percent']= laliga_df['Winning Percent'].fillna(0)
#print(laliga_df.iloc[0:61,[1,20]])
laliga_df.sort_values(by='Winning Percent',ascending=False,inplace=True)
print(laliga_df.iloc[0:5,[1,20]])
```

```python
# Group teams based on their “Best position” and print the sum of their points for all positions (7.5 points)
bestpos_df = pd.DataFrame(laliga_df.groupby('BestPosition')['Points'].agg('sum'))
```
